STEP 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

STEP:2 Load Dataset

```python
df = pd.read_csv("/kaggle/input/datasets/soumyasahoo23/restrruant-data/Dataset .csv")
df.head()
```

STEP 3: Preprocess Data

```python
df.fillna('', inplace=True)

# Clean column names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()

# Create combined feature column
df['combined_features'] = df['cuisines'].astype(str) + " " + df['price range'].astype(str)
```

STEP 4: Vectorize Text Data

```python
vectorizer = TfidfVectorizer()
feature_matrix = vectorizer.fit_transform(df['combined_features'])
```

STEP 5: Recommendation Function

```python
def recommend_restaurants(user_cuisine, user_price_range, top_n=5):
    
    
    user_input = f"{user_cuisine} {user_price_range}"
    user_vector = vectorizer.transform([user_input])
    similarities = cosine_similarity(user_vector, feature_matrix).flatten()
    similar_indices = similarities.argsort()[-top_n:][::-1]
    recommended = df.iloc[similar_indices][['restaurant name', 'cuisines', 'price range', 'aggregate rating']]
    
    return recommended
```

STEP 6: Test Recommendation

```python
user_cuisine = "Italian"
user_price_range = "2"

recommend_restaurants(user_cuisine, user_price_range)
```
